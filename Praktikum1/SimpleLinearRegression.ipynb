Import Library & Buat Data Dummy

```python
import pandas as pd

data = {
    'luas': [6, 8, 10, 14, 18],
    'harga': [7, 9, 13, 17.5, 18]
}

df = pd.DataFrame(data)
df
```

Visualisasi Scatter Plot

```python
import matplotlib.pyplot as plt

df.plot(kind='scatter', x='luas', y='harga')
plt.title('Perbandingan Luas Tanah dengan Harga Tanah')
plt.xlabel('Luas Tanah')
plt.ylabel('Harga Tanah')
plt.grid(True)
plt.show()
```

Pisahkan X dan y

```python
import numpy as np

X = np.array(df[['luas']])
y = np.array(df['harga'])
```

Latih Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)

print("Intercept:", model.intercept_)
print("Slope:", model.coef_)
```

Visualisasi Garis Regresi

```python
X_vis = np.array([0, 25]).reshape(-1,1)
y_vis = model.predict(X_vis)

plt.scatter(X, y)
plt.plot(X_vis, y_vis, color='red')
plt.show()
```

Prediksi Luas 12, 15, 20

```python
luas_tanah = np.array([12,15,20]).reshape(-1,1)
harga_prediksi = model.predict(luas_tanah)

for luas, hrg in zip(luas_tanah, harga_prediksi):
    print(f"Luas {luas[0]} → Prediksi harga: {hrg}")
```
